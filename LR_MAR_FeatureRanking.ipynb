```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Load CSV
df = pd.read_csv("Logistic Regression/outputs/logisticRegression_MAR_feature_importance.csv")

# Get unique MAR levels sorted
mar_levels = sorted(df["MAR_Level"].unique())

# Grid layout
n = len(mar_levels)
cols = 3
rows = math.ceil(n / cols)

# Create figure
fig, axes = plt.subplots(rows, cols, figsize=(18, 12))
axes = axes.flatten()

# Plot each MAR level
for i, mar in enumerate(mar_levels):
    subset = df[df["MAR_Level"] == mar].sort_values("Importance", ascending=False)
    
    axes[i].bar(subset["Feature"], subset["Importance"])
    axes[i].set_title(f"MAR = {mar}%")
    axes[i].set_ylabel("Mean |SHAP|")
    axes[i].tick_params(axis='x', rotation=45)

# Remove empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
