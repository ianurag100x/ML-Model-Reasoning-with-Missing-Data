```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# LOAD CSV FILES
# -----------------------------
lr_stability = pd.read_csv("Logistic Regression/outputs/logisticRegression_MAR_shap_stability.csv")
rf_stability = pd.read_csv("Random Forest/outputs/randomForest_MAR_shap_stability.csv")

# -----------------------------
# EXTRACT DATA
# -----------------------------
# Logistic Regression
lr_x = lr_stability["MAR_Level"].tolist()
lr_y = lr_stability["Spearman_Correlation"].tolist()

# Random Forest
rf_x = rf_stability["MAR_Level"].tolist()
rf_y = rf_stability["Spearman_Correlation"].tolist()

# -----------------------------
# PLOT
# -----------------------------
plt.figure(figsize=(10,6))

plt.plot(lr_x, lr_y, marker='o', linewidth=2, label="Logistic Regression")
plt.plot(rf_x, rf_y, marker='s', linewidth=2, label="Random Forest")

# Labels
plt.xlabel("MAR Level (%)")
plt.ylabel("Spearman Correlation")
plt.title("SHAP Stability vs MAR Level")
plt.xticks(sorted(set(lr_x + rf_x)))
plt.ylim(0, 1.05)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.tight_layout()
plt.show()
```
