```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# LOAD CSV FILES
# -----------------------------
lr_base = pd.read_csv("Logistic Regression/outputs/baseline_logisticRegression_metrics.csv")
lr_mar  = pd.read_csv("Logistic Regression/outputs/logisticRegression_MAR_results.csv")

rf_base = pd.read_csv("Random Forest/outputs/baseline_randomForest_metrics.csv")
rf_mar  = pd.read_csv("Random Forest/outputs/randomForest_MAR_results.csv")

# -----------------------------
# EXTRACT ACCURACY VALUES
# -----------------------------
# Logistic Regression
lr_x = [0] + lr_mar["MAR_Level"].tolist()
lr_y = [lr_base["Accuracy"].iloc[0]] + lr_mar["Accuracy"].tolist()

# Random Forest
rf_x = [0] + rf_mar["MAR_Level"].tolist()
rf_y = [rf_base["Accuracy"].iloc[0]] + rf_mar["Accuracy"].tolist()

# -----------------------------
# PLOT
# -----------------------------
plt.figure(figsize=(10,6))

plt.plot(lr_x, lr_y, marker='o', linewidth=2, label="Logistic Regression")
plt.plot(rf_x, rf_y, marker='s', linewidth=2, label="Random Forest")

# Labels
plt.xlabel("MCAR Level (%)")
plt.ylabel("Accuracy")
plt.title("Accuracy vs MCAR Level")
plt.xticks(lr_x)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.tight_layout()
plt.show()
```
