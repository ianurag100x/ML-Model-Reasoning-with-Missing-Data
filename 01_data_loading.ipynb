```python
# ==========================================================
# 01_data_loading.py
# Paper 2:
# The Impact of Missing Data on Model Reasoning:
# An Interpretability Stability Study Using Explainable AI
# ==========================================================

import pandas as pd
import numpy as np
import os

# ----------------------------------------------------------
# 1. Load Dataset
# ----------------------------------------------------------
file_path = "dataset/dengue_diseases.csv"

df = pd.read_csv(file_path)

print("=" * 60)
print("DATASET LOADED SUCCESSFULLY")
print("=" * 60)

# ----------------------------------------------------------
# 2. Basic Inspection
# ----------------------------------------------------------
print("\nFirst 5 Rows:\n")
print(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)
```

```python
# ----------------------------------------------------------
# 3. Missing Values Check (Original Dataset)
# ----------------------------------------------------------
print("\nOriginal Missing Values:")
print(df.isnull().sum())

# ----------------------------------------------------------
# 4. Duplicate Row Check
# ----------------------------------------------------------
duplicates = df.duplicated().sum()
print(f"\nDuplicate Rows Found: {duplicates}")

if duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicate rows removed.")
```

```python
# ----------------------------------------------------------
# 5. Encode Gender Column
# ----------------------------------------------------------
print("\nUnique Gender Values:")
print(df["gender"].unique())

gender_mapping = {
    "Male": 0,
    "Female": 1,
    "Child": 2
}

df["gender"] = df["gender"].map(gender_mapping)

# ----------------------------------------------------------
# 6. Ensure Numeric Types
# ----------------------------------------------------------
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# ----------------------------------------------------------
# 7. Final Missing Check After Conversion
# ----------------------------------------------------------
print("\nMissing Values After Numeric Conversion:")
print(df.isnull().sum())

# ----------------------------------------------------------
# 8. Separate Features and Target
# ----------------------------------------------------------
target_column = "dengue_label"

X = df.drop(columns=[target_column])
y = df[target_column]

print("\nFeature Matrix Shape:", X.shape)
print("Target Vector Shape :", y.shape)

# ----------------------------------------------------------
# 9. Class Distribution
# ----------------------------------------------------------
print("\nClass Distribution:")
print(y.value_counts())

print("\nClass Percentage:")
print((y.value_counts(normalize=True) * 100).round(2))
```

```python
# ----------------------------------------------------------
# 10. Statistical Summary
# ----------------------------------------------------------
print("\nFeature Summary:")
print(X.describe())

# ----------------------------------------------------------
# 11. Save Clean Reference Dataset
# ----------------------------------------------------------
os.makedirs("outputs", exist_ok=True)

save_path = "outputs/clean_dengue_dataset.csv"
df.to_csv(save_path, index=False)

print(f"\nClean dataset saved to: {save_path}")

print("\nPROCESS COMPLETED SUCCESSFULLY")
```
