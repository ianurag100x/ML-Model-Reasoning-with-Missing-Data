```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# LOAD CSV FILES
# -----------------------------
lr_acc = pd.read_csv("Logistic Regression/outputs/logisticRegression_MAR_results.csv")
lr_rank = pd.read_csv("Logistic Regression/outputs/logisticRegression_MAR_shap_stability.csv")

rf_acc = pd.read_csv("Random Forest/outputs/randomForest_MAR_results.csv")
rf_rank = pd.read_csv("Random Forest/outputs/randomForest_MAR_shap_stability.csv")

# -----------------------------
# EXTRACT DATA
# -----------------------------
x = sorted(lr_acc["MAR_Level"].unique())

# Logistic Regression
lr_accuracy = lr_acc["Accuracy"].tolist()
lr_spearman = lr_rank["Spearman_Correlation"].tolist()

# Random Forest
rf_accuracy = rf_acc["Accuracy"].tolist()
rf_spearman = rf_rank["Spearman_Correlation"].tolist()

# -----------------------------
# PLOT
# -----------------------------
plt.figure(figsize=(11,6))

plt.plot(x, lr_accuracy, marker='o', linewidth=2, label="LR_Accuracy")
plt.plot(x, lr_spearman, marker='o', linestyle='--', linewidth=2, label="LR_Rank")

plt.plot(x, rf_accuracy, marker='s', linewidth=2, label="RF_Accuracy")
plt.plot(x, rf_spearman, marker='s', linestyle='--', linewidth=2, label="RF_Rank")

# Labels
plt.xlabel("MAR Level (%)")
plt.ylabel("Score")
plt.title("Accuracy vs Interpretability Stability Across MAR Levels")
plt.xticks(x)
plt.ylim(0, 1.05)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(loc="lower right")

plt.tight_layout()
plt.show()
```
